```python
"""
read wav files return sample,sample_rate
"""
def read_wav(filename='/yes/0a7c2a8d_nohash_0.wav',train_audio_path = '',sr=None):
  
   samples,sample_rate = librosa.load(str(train_audio_path) + filename,sr=sr)
   return samples,sample_rate

"""
return audio player in the notebook when called
"""
def listen(samples,sample_rate,):
  return ipd.Audio(samples, rate=sample_rate)

"""
iterate path to load all audio file
"""
def iter_list(train_audio_path = 'train/audio/'):
  sample_list=[]
  for filename in os.listdir(train_audio_path):
    sample_rate,sample=read_wav(filename=filename,train_audio_path=train_audio_path)
    sample_list.append(np.array([sample_rate,sample]))
  return np.array(sample_list)
```

* Signal Plots

```python
def plot_signal(signals):
    fig, axes = plt.subplots(nrows=1, ncols=1, sharex=False,
                             sharey=True, figsize=(20,5))
    axes.set_title("sig")
    axes.plot(list(signals))
    #axes.get_xaxis().set_visible(False)
    #axes.get_yaxis().set_visible(False)
    
def plot_signals(signals):
    fig, axes = plt.subplots(nrows=2, ncols=5, sharex=False,
                             sharey=True, figsize=(20,5))
    fig.suptitle('Time Series', size=16)
    i = 0
    for x in range(2):
        for y in range(5):
            axes[x,y].set_title(list(signals.keys())[i])
            axes[x,y].plot(list(signals.values())[i])
            axes[x,y].get_xaxis().set_visible(False)
            axes[x,y].get_yaxis().set_visible(False)
            i += 1

def plot_fft(fft):
    fig, axes = plt.subplots(nrows=2, ncols=5, sharex=False,
                             sharey=True, figsize=(20,5))
    fig.suptitle('Fourier Transforms', size=16)
    i = 0
    for x in range(2):
        for y in range(5):
            data = list(fft.values())[i]
            Y, freq = data[0], data[1]
            axes[x,y].set_title(list(fft.keys())[i])
            axes[x,y].plot(freq, Y)
            axes[x,y].get_xaxis().set_visible(False)
            axes[x,y].get_yaxis().set_visible(False)
            i += 1

def plot_fbank(fbank):
    fig, axes = plt.subplots(nrows=2, ncols=5, sharex=False,
                             sharey=True, figsize=(20,5))
    fig.suptitle('Filter Bank Coefficients', size=16)
    i = 0
    for x in range(2):
        for y in range(5):
            axes[x,y].set_title(list(fbank.keys())[i])
            axes[x,y].imshow(list(fbank.values())[i],
                    cmap='hot', interpolation='nearest')
            axes[x,y].get_xaxis().set_visible(False)
            axes[x,y].get_yaxis().set_visible(False)
            i += 1

def plot_mfccs(mfccs):
    fig, axes = plt.subplots(nrows=2, ncols=5, sharex=False,
                             sharey=True, figsize=(20,5))
    fig.suptitle('Mel Frequency Cepstrum Coefficients', size=16)
    i = 0
    for x in range(2):
        for y in range(5):
            axes[x,y].set_title(list(mfccs.keys())[i])
            axes[x,y].imshow(list(mfccs.values())[i],
                    cmap='hot', interpolation='nearest')
            axes[x,y].get_xaxis().set_visible(False)
            axes[x,y].get_yaxis().set_visible(False)
            i += 1
```

# Reading datasets

## SAVE dataset

```python
def SAVE(path=SAVEE):
  dir_list = os.listdir(path)

  # parse the filename to get the emotions
  emotion=[]
  path = []
  for i in dir_list:
      if i[-8:-6]=='_a':
          emotion.append('male_angry')
      elif i[-8:-6]=='_d':
          emotion.append('male_disgust')
      elif i[-8:-6]=='_f':
          emotion.append('male_fear')
      elif i[-8:-6]=='_h':
          emotion.append('male_happy')
      elif i[-8:-6]=='_n':
          emotion.append('male_neutral')
      elif i[-8:-6]=='sa':
          emotion.append('male_sad')
      elif i[-8:-6]=='su':
          emotion.append('male_surprise')
      else:
          emotion.append('male_error') 
      path.append(SAVEE +"/"+ i)
      
  # Now check out the label count distribution 
  SAVEE_df = pd.DataFrame(emotion, columns = ['labels'])
  #SAVEE_df['source'] = 'SAVEE'
  SAVEE_df = pd.concat([SAVEE_df, pd.DataFrame(path, columns = ['path'])], axis = 1)
  #SAVEE_df.labels.value_counts()
  return SAVEE_df
```

## RAV dataset

```python
def RAVV(path=RAV):
  dir_list = os.listdir(path)
  dir_list.sort()

  emotion = []
  gender = []
  path_list = []
  for i in dir_list:
      fname = os.listdir(path +"/"+ i)
      for f in fname:
          part = f.split('.')[0].split('-')
          emotion.append(int(part[2]))
          temp = int(part[6])
          if temp%2 == 0:
              temp = "female"
          else:
              temp = "male"
          gender.append(temp)
          path_list.append(path +"/"+ i + '/' + f)

          
  RAV_df = pd.DataFrame(emotion)
  RAV_df = RAV_df.replace({1:'neutral', 2:'neutral', 3:'happy', 4:'sad', 5:'angry', 6:'fear', 7:'disgust', 8:'surprise'})
  RAV_df = pd.concat([pd.DataFrame(gender),RAV_df],axis=1)
  RAV_df.columns = ['gender','emotion']
  RAV_df['labels'] =RAV_df.gender + '_' + RAV_df.emotion
  #RAV_df['source'] = 'RAVDESS'  
  RAV_df = pd.concat([RAV_df,pd.DataFrame(path_list, columns = ['path'])],axis=1)
  RAV_df = RAV_df.drop(['gender', 'emotion'], axis=1)
  #RAV_df.labels.value_counts()
  return RAV_df
```

## TESS dataset

```python
def TESSS(path=TESS):
  path_list = []
  emotion = []
  dir_list = os.listdir(path)
  dir_list.sort()

  for i in dir_list:
      fname = os.listdir(path + "/"+i)
      for f in fname:
          if i == 'OAF_angry' or i == 'YAF_angry':
              emotion.append('female_angry')
          elif i == 'OAF_disgust' or i == 'YAF_disgust':
              emotion.append('female_disgust')
          elif i == 'OAF_Fear' or i == 'YAF_fear':
              emotion.append('female_fear')
          elif i == 'OAF_happy' or i == 'YAF_happy':
              emotion.append('female_happy')
          elif i == 'OAF_neutral' or i == 'YAF_neutral':
              emotion.append('female_neutral')                                
          elif i == 'OAF_Pleasant_surprise' or i == 'YAF_pleasant_surprised':
              emotion.append('female_surprise')               
          elif i == 'OAF_Sad' or i == 'YAF_sad':
              emotion.append('female_sad')
          else:
              emotion.append('Unknown')
          path_list.append(path+"/" + i + "/" + f)

  TESS_df = pd.DataFrame(emotion, columns = ['labels'])
  #TESS_df['source'] = 'TESS'
  TESS_df = pd.concat([TESS_df,pd.DataFrame(path_list, columns = ['path'])],axis=1)
  #TESS_df.labels.value_counts()
  return TESS_df
```

# Removing Noise

## enveloppe

```python
"""
this enveloppe remove low volume voices
"""
def enveloppe(sig,sr,threshhold):
  mask=[]
  sig=pd.Series(sig).apply(np.abs)
  sig_mean=sig.rolling(window=int(sr/10),min_periods=1,center=True).mean()
  for mean in sig_mean:
    if mean>threshhold:
      mask.append(True)
    else:  mask.append(False)
  return mask
```

```python
def test_enveloppe(filepath="/content/dataset/SAVEE/ALL/DC_a01.wav",thresh=0.0005,print_mask=True):
  sample,sr=read_wav(filepath,sr=44000)
  plot_signal(sample)
  mask=enveloppe(sample,sr,thresh)
  print(len(mask),"  :  ",len(sample))
  plot_signal(sample[mask])
  if(print_mask):  return listen(sample[mask],sr)
  else : return listen(sample,sr)
```

```python
test_enveloppe(thresh=0.02,print_mask=True)
```

## add  simple noise

```python
def addNoise(samples,sample_rate,coef):
  # Set a target channel noise power to something very noisy
  target_noise_db = 10

  # Convert to linear Watt units
  target_noise_watts = 10 ** (target_noise_db / 10)

  # Generate noise samples
  mean_noise = 0
  noise_volts = np.random.normal(mean_noise, np.sqrt(target_noise_watts), len(samples))

  # Noise up the original signal (again) and plot
  y_volts = samples + noise_volts*coef
  return y_volts
```

```python
def testaddNoise(coef,filepath="/content/dataset/SAVEE/ALL/DC_a01.wav",print_mask=False):
  sample,sr=read_wav(filepath)
  plot_signal(sample)
  sample2=addNoise(sample,sr,coef)
  plot_signal(sample2)
  if(print_mask):  return listen(sample2,sr)
  else : return listen(sample,sr)
```

## Gaussian Noise

```python
def Gauss_noise(k,sig):
     return np.random.normal(scale=k*np.max(sig), size=len(sig))

def GN(samples,sample_rate,k=2e-2):
    noise_g = Gauss_noise(k,samples)
    return samples+noise_g

    
```

```python
def test_transform(filepath="/content/dataset/SAVEE/ALL/DC_a01.wav",print_mask=False):
  sample,sr=read_wav(filepath,sr=16000)
  plot_signal(sample)
  
  sample2=GN(sample,sr)
  plot_signal(sample2)
  if(print_mask):  return listen(sample2,sr)
  else : return listen(sample,sr)


```

## add pitch and speed

```python
def pitch_and_speed(samples,sample_rate) :
  y_pitch_speed = samples.copy()
  # you can change low and high here
  length_change = np.random.uniform(low=0.8, high = 1)
  speed_fac = 1.0  / length_change
  #print("resample length_change = ",length_change)
  tmp = np.interp(np.arange(0,len(y_pitch_speed),speed_fac),np.arange(0,len(y_pitch_speed)),y_pitch_speed)
  minlen = min(y_pitch_speed.shape[0], tmp.shape[0])
  y_pitch_speed *= 0
  y_pitch_speed[0:minlen] = tmp[0:minlen]
  #Audio(y_pitch_speed, rate=sample_rate)
  return y_pitch_speed

```

## change pitch

```python
def pitch(samples,sample_rate ):
  y_pitch = samples.copy()
  bins_per_octave = 12
  pitch_pm = 2
  pitch_change =  pitch_pm * 2*(np.random.uniform())   
  #print("pitch_change = ",pitch_change)
  y_pitch = librosa.effects.pitch_shift(y_pitch.astype('float64'), 
                                        sample_rate, n_steps=pitch_change, 
                                        bins_per_octave=bins_per_octave)
  #Audio(y_pitch, rate=sample_rate)
  return y_pitch
```

## change speed

```python
def speed(samples,sample_rate):
  y_speed = samples.copy()
  speed_change = np.random.uniform(low=0.9,high=1.1)
  #print("speed_change = ",speed_change)
  tmp = librosa.effects.time_stretch(y_speed.astype('float64'), speed_change)
  minlen = min(y_speed.shape[0], tmp.shape[0])
  y_speed *= 0 
  y_speed[0:minlen] = tmp[0:minlen]
  #Audio(y_speed, rate=sample_rate)
  return y_speed
```

## value aug

```python
def value_aug(samples,sample_rate):
  y_aug = samples.copy()
  dyn_change = np.random.uniform(low=1.5,high=3)
  #print("dyn_change = ",dyn_change)
  y_aug = y_aug * dyn_change
  return y_aug
```

## noise dist

```python
def noise_dist(samples,sample_rate):
  y_noise = samples.copy()
  # you can take any distribution from https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.random.html
  noise_amp = 0.005*np.random.uniform()*np.amax(y_noise)
  y_noise = y_noise.astype('float64') + noise_amp * np.random.normal(size=y_noise.shape[0])
  #Audio(y_noise, rate=sample_rate)
  return y_noise
```

## random shifting

```python
def rand_shift(samples,sample_rate):
  y_shift = samples.copy()
  timeshift_fac = 0.2 *2*(np.random.uniform()-0.5)  # up to 20% of length
  #print("timeshift_fac = ",timeshift_fac)
  start = int(y_shift.shape[0] * timeshift_fac)
  #print(start)
  if (start > 0):
      y_shift = np.pad(y_shift,(start,0),mode='constant')[0:y_shift.shape[0]]
  else:
      y_shift = np.pad(y_shift,(0,-start),mode='constant')[0:y_shift.shape[0]]
  #Audio(y_shift, rate=sample_rate)
  return y_shift
```

## Stretching

```python
def stretching(samples,sample_rate):
  input_length = len(samples)
  streching = samples.copy()
  streching = librosa.effects.time_stretch(streching.astype('float'), 1.1)
  if len(streching) > input_length:
      streching = streching[:input_length]
  else:
      streching = np.pad(streching, (0, max(0, input_length - len(streching))), "constant")
  #Audio(streching, rate=sample_rate)
  return streching
```

```python
#Used features
features_col=["mfcc","rmse","chroma","spec_cent","spec_bw","rolloff","zcr"]
```

```python
def create_dataset_2D(XX):
  global X
  all_res=[] #final result
  to_keep=[]
  for index,path in enumerate(XX["path"]):
      res={} #row result
      sample, sample_rate = librosa.load(path
                                    #, res_type='kaiser_fast'
                                    #,duration=2.5
                                    #,sr=16000
                                    #,offset=0
                                  )
      #calculate the enveloppe
      mask=enveloppe(sample,sample_rate,0.0001)

      sample_rate = np.array(sample_rate)
      #if enveloppe removed everything then we skip the row
      if (sample[mask].shape[0]==0):
        continue
      #index to keep
      to_keep.append(index)
      #apply the enveloppe
      sample=sample[mask]
      #apply data augmentation (choose 1 each time)
      sample=pitch(sample,sample_rate)

      #calculate log mel no longer used
      #mel=librosa.feature.melspectrogram(y=XX[mask],sr=sample_rate, n_mels=128,n_fft = 128 * 20)
      #log_mel=librosa.power_to_db(mel, ref=np.max)


      #extracting features
      mfccs = librosa.feature.mfcc(y=sample,sr=sample_rate,n_mfcc=20)
      rmse = librosa.feature.rmse(y=sample)
      chroma_stft = librosa.feature.chroma_stft(y=sample,sr=sample_rate)
      spec_cent = librosa.feature.spectral_centroid(y=sample,sr=sample_rate)
      spec_bw = librosa.feature.spectral_bandwidth(y=sample,sr=sample_rate)
      rolloff = librosa.feature.spectral_rolloff(y=sample,sr=sample_rate)
      zcr = librosa.feature.zero_crossing_rate(y=sample)
      
      res["mfcc"]=mfccs
      res["rmse"]=rmse
      res["chroma"]=chroma_stft
      res["spec_cent"]=spec_cent
      res["spec_bw"]=spec_bw
      res["rolloff"]=rolloff
      res["zcr"]=zcr
      #res["mel"]=mel
      #res["log_mel"]=log_mel

      #add Row to the final result
      all_res.append(res)
      #add timestep
      X.iloc[index,2]=mfccs.shape[1]
      #add length
      X.iloc[index,3]=XX.shape[0]/sample_rate
  #return kept indices and the final result (kept indices are needed for saving the files)    
  return to_keep,all_res
```

```python
keep_ind,X_2d=create_dataset_2D(X)
```
