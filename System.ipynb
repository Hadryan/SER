# Adjust Dataset

* Dataset Need to be adjusted because keras functional api multi input needs inputs in a table which are the features in our case

original shape: (row,feature,**feature)


result shape: (feature_number,row,**feature)

```python
def adjust_dataset(X=X_train,max_timestep=80):
  X_t=[]
  for ind,feat in enumerate(features_col):
    X_s=[]
    for temp in X:
      X_s.append(temp[feat])
    X_t.append(X_s)


  for i in range(len(X_t)):
    for j in range(len(X_t[i])):
      X_t[i][j]=np.pad( X_t[i][j],((0,max(0,max_timestep-X_t[i][j].shape[0])),(0,0)),mode="constant",constant_values=((0, 0),(0, 0)))[:max_timestep,:]
    X_t[i]=np.array(X_t[i])
    #X_t[i]=np.expand_dims(X_t[i], axis=-1)
  
  
  return X_t
```

```python
X_train=adjust_dataset(X=X_train)
X_test=adjust_dataset(X=X_test)
```

```python
# Functions for visualizing the training parameters (loss and accuracy)
def loss_plot(fit_history):
    plt.figure(figsize=(18, 4))
    plt.plot(fit_history.history['loss'], label = 'train')
    plt.plot(fit_history.history['val_loss'], label = 'test')
    plt.legend()
    plt.title('Loss Function');
    
def accuracy_plot(fit_history):
    plt.figure(figsize=(18, 4))
    plt.plot(fit_history.history['accuracy'], label = 'train')
    plt.plot(fit_history.history['val_accuracy'], label = 'test')
    plt.legend()
    plt.title('Accuracy');
```

* one part of time distributed convolution network (dimension: 1)

```python
def tdc1(filters=128,input=input,name="test"):
  y = TimeDistributed(Conv1D(filters, kernel_size=( 3), strides=( 1), padding='same'), name='Conv_1_'+name)(input)
  y = TimeDistributed(BatchNormalization(), name='BatchNorm_'+name)(y)
  y = TimeDistributed(Activation('elu'), name='Activ_'+name)(y)
  y = TimeDistributed(MaxPooling1D(pool_size=( 2), strides=( 2), padding='same'), name='MaxPool_1_'+name)(y)
  y = TimeDistributed(Dropout(0.1), name='Drop_'+name)(y)
  return y
```

* one part of time distributed convolution network (dimension: 2) (not used)

```python
def tdc2(filters=128,input=input,name="test"):
  y = TimeDistributed(Conv2D(filters, kernel_size=( 3), strides=( 1), padding='same'), name='Conv_1_'+name)(input)
  #y = TimeDistributed(BatchNormalization(), name='BatchNorm_'+name)(y)
  y = TimeDistributed(Activation('elu'), name='Activ_'+name)(y)
  y = TimeDistributed(MaxPooling2D(pool_size=( 2), strides=( 2), padding='same'), name='MaxPool_1_'+name)(y)
  y = TimeDistributed(Dropout(0.1), name='Drop_'+name)(y)
  return y
```

## System Modeling

## Model architecture

```python
def sys(X_train,y_train,path=None,features_col=features_col):
  input={}
  y={}
  for feat in features_col:
    if (feat=="mfcc"):
      input[feat] = Input(shape=(None,20,1), name='Input_'+feat)
      y[feat]=tdc1(filters=64,input=input[feat],name=feat+"2")
      y[feat]=tdc1(filters=64,input=y[feat],name=feat+"4")
      y[feat]=tdc1(filters=64,input=y[feat],name=feat+"7")
      y[feat]=tdc1(filters=64,input=y[feat],name=feat+"8")
      y[feat]=tdc1(filters=64,input=y[feat],name=feat+"9")
      y[feat] = TimeDistributed(Flatten(), name=TN
2
'Flat_'+feat)(y[feat])   
      y[feat] = LSTM(32, return_sequences=False, dropout=0.1, name='LSTM_2_'+feat)(y[feat])
      attention_probs = Dense( 32, activation='sigmoid', name='attention_probs1_'+feat)(y[feat])
      attention_mul = multiply([ attention_probs, y[feat]], name='attention_mul1_'+feat)
      y[feat] = Dense(32, activation='softmax', name='FC_'+feat)(attention_mul)
    else :
      input[feat] = Input(shape=(None,1,1), name='Input_'+feat)
      y[feat]=tdc1(filters=16,input=input[feat],name=feat+"1")
      y[feat] = TimeDistributed(Flatten(), name='Flat_'+feat)(y[feat])
      y[feat] = LSTM(16, return_sequences=False, dropout=0.1, name='LSTM_2_'+feat)(y[feat])
     
  res1=Concatenate(axis=-1)([y[feat] for feat in features_col if feat!="mfc"])
  res=Dense(y_train.shape[1],activation='softmax', name='FC_last')(res1)

  # Build final model
  model = Model(inputs=[input[feat] for feat in features_col], outputs=res)
  # Compile model
  if (path!=None):
        model=load_model(path)
        #model.load_weights(path)
  model.compile(optimizer="adam", loss='categorical_crossentropy', metrics=['accuracy'])
  model.summary()
  return model
```
